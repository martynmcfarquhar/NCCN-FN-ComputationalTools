# Installing MATLAB and SPM
As mentioned in the previous section, SPM is the software we will be using to analyse brain images on this course. SPM is developed by the [Wellcome Centre for Human Neuroimaging](https://www.fil.ion.ucl.ac.uk/) at University College London (UCL) and is notable for being written in the MATLAB programming language. This means that we need a copy of MATLAB in order for SPM to run. As such, our first job is to get MATLAB up and running.

## MATLAB
MATLAB (MATrix LABoratory) is both a programming language *and* a scientific computing environment, developed by the [MathWorks](https://uk.mathworks.com/) corporation. It was originally developed as a simple interactive matrix calculator, but has since expanded into a fully-fledged programming language with suites of functions for image processing, computer vision, statistics, machine learning and the development of graphical user interfaces (to name a few). In this section, we will see how to install MATLAB and have a quick tour of the interface. We will then start learning how to use MATLAB in the next section.

### Installing MATLAB
Despite its roots in academia, MATLAB is no longer free software. If you are using a University cluster PC, MATLAB will already be installed and ready for you to use. However, if you want to use your own computer, you will need to install MATLAB with the University campus-wide license. To do this, first, visit the [MathWorks website](https://uk.mathworks.com/) and click the <kbd>Get MATLAB</kbd> button. You will be prompted to create an account. Make sure you do this using your University email address, otherwise the license will not be available. Once you have signed in, click the <kbd>Install MATLAB</kbd> button. Once you are on the downloads page, you can select the version that you want to install. For Windows, just install the latest version. If you are using a Mac, make sure you install version `R2023a`, otherwise there will be compatibility issues with SPM12.

## SPM
Now that MATLAB is installed, we can turn our attention to the SPM analysis software. SPM is one of the oldest and most widely used software packages for processing brain images. It was originally developed in the early 90s by [Karl Friston](https://en.wikipedia.org/wiki/Karl_J._Friston), as a means of processing and analysing PET data. With the advant of fMRI, the software was soon expanded for use with functional imaging data, as well as structural MRI, MEG and EEG datasets. The version of SPM we will be using is SPM12, which is the 5th major version of the SPM software and was released at the end of 2012. SPM is a hugely import piece of academic software. The SPM team have been highly influential in the field of fMRI data analysis, developing many of the methods and approaches we will be learning about on this course. For those interested in learning more, much of the history of SPM has been documented in the retrospective by [Ashburner (2012)](https://doi.org/10.1016/j.neuroimage.2011.10.025). 

### Installing SPM
At its most basic, SPM is a collection of MATLAB functions (we will discuss MATLAB functions a little later). You download the software as a folder, tell MATLAB where that folder is located and then launch the software by typing `spm` at the MATLAB prompt. The SPM launcher will then open, where you can select which flavour of SPM to open, as shown below.

````matlab
spm
````

In order to install SPM, follow along with the video below. Note that if you are using macOS, there will be some additional steps to perform to get SPM working, which are explained in the section below the video.

### Resolving Errors on macOS
For anyone installing SPM on macOS, there are some additional steps that will be required before the software will function correctly. If you try and launch SPM, you will probably get some errors in MATLAB complaining about the MEX files. The issue is that the macOS security system is rather over-zealous and will have quarantined the MEX files on the off-chance that they contain malicious code. In order to resolve this, we have to un-quarantine those files. Unfortunately, Apple do not want to make it too easy to bypass the security systems. So, we need to type some commands into the macOS Terminal.

To begin with, you need to open the Terminal application. It can be found inside `Applications > Utilities`, or you can press <kbd>CMD</kbd> + <kbd>SPACE</kbd> to bring up Spotlight search, and then type `terminal`. Once open, you need to move the current directory the terminal is working inside to the `spm12` folder. Assuming you installed SPM inside `Documents > MATLAB`, you can use the `cd` command to move there by typing

```bash
cd ~/Documents/MATLAB/spm12
```

and hitting <kbd>Enter</kbd>. Then, you need to type the following command

```bash
find . -name "*.mexmaci64" -exec xattr -d com.apple.quarantine {} \;
```

and hit <kbd>Enter</kbd>. This will go through the `spm12` directory and un-quarantine all the macOS-specific MEX files. Once that has finished, close the Terminal and then re-open MATLAB and try relaunching SPM. With any luck, the errors about MEX files will have disappeared. However, if you are still have issues, see the next section.

### Resolving Other Problems
One of the first places to look if you having issues getting SPM working is the [SPM Wiki](https://en.wikibooks.org/wiki/SPM). This is a very useful technical resource with guides on many different aspects of SPM, including installation and custom compliation of the MEX files (which hopefully you never have to do!). Another resource is the [SPM JISCM@il](https://www.jiscmail.ac.uk/cgi-bin/webadmin?A0=SPM). This is a question-and-answer forum that both the users and developers of SPM frequent to discuss any problems or give advice on using the software. If you are having issues there is a good chance that someone else may have asked a question about it already on the list. So try searching the archives or creating a new post to find help with any problems. If you still cannot resolve the problem on your own, please post on the relevant Blackboard discussion board for the current modules. One of the teaching team, or even your fellow students, can then respond to try and help.
